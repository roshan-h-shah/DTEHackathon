```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, classification_report

# Define the parameter grid you want to search
param_grid = {
    'C': [0.1, 1, 10, 100, 1000],              # Different values for C (regularization parameter)
    'kernel': [ 'poly', 'rbf',],  # Different kernel types
    'gamma': ['scale'],                 # Different values for gamma
    'degree': [2, 3, 4, 5]                      # Degrees for the polynomial kernel (only used if kernel='poly')
}

# Initialize SVC
svc = SVC()

# Use GridSearchCV to search over the parameter grid
grid_search = GridSearchCV(estimator=svc, param_grid=param_grid, cv=5, scoring='accuracy', verbose=2, n_jobs=-1)

# Train the model using the best parameters found by GridSearchCV
grid_search.fit(X_train, y_train)

# Get the best parameters and estimator
best_params = grid_search.best_params_
best_svc = grid_search.best_estimator_

# Predict using the best found model
y_pred = best_svc.predict(X_test)

# Print accuracy and classification report
print("Best Parameters:", best_params)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
